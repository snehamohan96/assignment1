Q13. How does the type of app affect its price? Create a bar chart to compare average prices by app type.

```python
#Type of app affect its price showing by plotting bar chart

google_df=google_df.dropna(subset=['Type','Price'])# remove missing values in the columns
```

```python
#google_df['Price']=google_df['Price'].str.replace('$','').astype(float)
avg_price_type = google_df.groupby('Type')['Price'].mean().reset_index()
```

Q14. What are the top 10 most popular apps in the dataset? Create a frequency table to identify the apps with the highest number of installs.



```python
#top 10 popular apps based on number of install

google_df = google_df.dropna(subset=['Installs'])#remove missing values in the column 'installs'
```

```python
#Remove non-numeric characters from the column and convert to numeric
google_df['Installs'] = google_df['Installs'].str.replace('[^0-9]', '', regex=True).astype(int)
```

```python
#grouping the data by 'App' and calculate number of installs for each app
app_installs = google_df.groupby('App')['Installs'].sum().reset_index()
```

```python
#sort the app by number of installs in descending order
top_10_apps = app_installs.sort_values(by='Installs', ascending=False).head(10)
```

```python
top_10_apps
```

```python
#create a frequency table of app installs
app_installs_freq = google_df['App'].value_counts().reset_index()
app_installs_freq.columns = ['App', 'Frequency']

#sort the frequency table by number of installs in descending order
app_installs_freq = app_installs_freq.sort_values(by='Frequency', ascending=False)
print(app_installs_freq.head(10))
```
