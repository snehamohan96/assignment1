```python
#Elbow Method for finding optimal number of clusters
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# Features used for clustering
features = ['Customer ID','Age','Total Spend','Items Purchased', 'Average Rating', 'Discount Applied','Days Since Last Purchase']
# Standardize the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df[features])
# Determine number of clusters using the elbow method
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=42)
    kmeans.fit(scaled_features)
    wcss.append(kmeans.inertia_)

# Plot the elbow graph
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss, marker='*',color='red')
plt.title('')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

# label encoders initialization
label_encoder_satisfaction = LabelEncoder()

# Encode 'Satisfaction Level' to numerical values
df['Satisfaction Level'] = label_encoder_satisfaction.fit_transform(df['Satisfaction Level'])

# one-hot encoding for categorical variables
df_encoded = pd.get_dummies(df, columns=['Gender', 'City', 'Membership Type', 'Discount Applied'])
```

```python
# Define target variable
X = df_encoded.drop(columns=['Customer ID', 'Satisfaction Level'])
y = df_encoded['Satisfaction Level']

# standardize the features
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

# train the model
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# predictions
y_pred = model.predict(X_test)

# model Evaluation
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
```
